#**Exploratory Data Analysis (EDA) and Visualization**

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x= "Acquisition_Cost" , y= "ROI" , data= df)
plt.title("Acquisition_Cost Vs ROI")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x= "Channel_Used", y= "Conversion_Rate" , hue= "Campaign_Type", data= df)
plt.title("Average Conversion Rate for different channel used by Campaign type")
plt.show()
```

```python
plt.figure()
sns.boxplot(x= "Campaign_Type" , y= "Engagement_Score" , data= df)
plt.title("Distribution of Engagement Score across different Campaign Type")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x= "Company" , y="ROI" , data= df)
plt.title("Average ROI by Company")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
correlation = df[['Engagement_Score' , 'Conversion_Rate']].corr()
sns.heatmap(correlation, annot= True)
plt.title("correlation between Engagement_Score and Conversion_Rate")
plt.show()
```

**Customer Segmentation**

```python
plt.figure(figsize=(8,6))
sns.countplot(x= "Target_Audience" , data= df)
plt.title("Distribution of Target_Audience")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data= df , x= 'Language' , y= 'Conversion_Rate' , hue= 'Customer_Segment')
plt.title("Highest Conversion Rate for each Language by Customer Segment")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x= "Customer_Segment" , y= "Acquisition_Cost" , data= df , hue= "Channel_Used")
plt.title("Distribution of Acquisition_Cost across Customer_Segment categorized by Channel_Used")
plt.xticks(rotation= 45)
plt.show()
```

```python
plt.figure()
sns.barplot(x= "Language" , y= "Conversion_Rate" , data= df)
plt.title("Language Vs Conversion Rate")
plt.show()
```

**Channel Effectivenes**

```python
plt.figure(figsize= (12,6))
sns.barplot(x= "Channel_Used" , y= "Engagement_Score" , hue= "Campaign_Type" , data= df)
plt.title("Engagement Score by Channel Used & Campaign Type")
plt.show()
```

```python
plt.figure()
roi_channel = df.groupby('Channel_Used')['ROI'].sum()
plt.pie(roi_channel , labels= roi_channel.index, autopct= '%1.1f%%')
plt.title('ROI vs Channel Used')
plt.show()
```

```python
plt.figure(figsize= (12,8))
sns.scatterplot(x= "Impressions" , y= "Clicks" , hue= "Channel_Used" , data= df)
plt.title("Relationship between Clicks and Impressions")
plt.show()
```

**Time Based Analysis**


```python
plt.figure()
plt.hist(df['Duration'] , bins= 20)
plt.title("Distribution of Duration")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst= True)
plt.figure(figsize= (20,8))
sns.lineplot(x= "Date" , y= "Conversion_Rate" , hue= "Company" , data= df , errorbar= None)
plt.title("Conversion Rate Vs Time")
plt.show()
```

```python
plt.figure(figsize= (8,4))
sns.lineplot(data=df, x='Date', y='Engagement_Score' , errorbar= None)
plt.title("Engagement Score Over Time")
plt.show()
```

**Geographic Analysis**

```python
plt.figure(figsize= (10,6))
sns.barplot(x= "Location" , y= "Acquisition_Cost" , data=df , hue= 'Location' , palette= 'pastel')
plt.title("Location Vs Acquisition_Cost")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize= (12,6))
sns.barplot(x= "Location" , y= "Conversion_Rate" , hue= "Target_Audience" , data= df)
plt.title("Conversion Rate Vs Location")
plt.show()
```

```python
roi_location = df.groupby("Location")['ROI'].sum()
plt.figure()
plt.pie(roi_location , labels= roi_location.index , autopct= '%1.1f%%')
plt.title("ROI Vs Location")
plt.show()
```
